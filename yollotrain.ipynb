```python
import os

from ultralytics import YOLO


model = YOLO('yolov8n-seg.pt')  # load a pretrained model (recommended for training)

model.train(data='/content/gdrive/My Drive/yolobuild/config.yaml', epochs=10, imgsz=640)
```

```python
model1= YOLO('/content/runs/segment/train/weights/best.pt')
```

```python
from PIL import Image
from ultralytics import YOLO
import tqdm, os

testfol = "/content/gdrive/My Drive/yolobuild/test/"

# Run inference on 'bus.jpg'
for i in tqdm.tqdm(os.listdir(testfol)):
  img = testfol+i
  results = model1(img)  # results list

  # Show the results
  for r in results:
      im_array = r.plot()  # plot a BGR numpy array of predictions
      im = Image.fromarray(im_array[..., ::-1])  # RGB PIL image
      im.save("/content/gdrive/My Drive/yolobuild/testres/"+i)  # save image

```
